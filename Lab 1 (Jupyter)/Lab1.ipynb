# Упражнение №1

Запишите выражение, заданное формулой, в виде, подходящем для языка Python.

$ \Large
y = \log_{1+x^2}{\frac{e^(\frac{1}{\sin(x)+1})}{\frac{5}{4} + \frac{1}{x^15}}} $


```python
x = [1, 10, 1000]
y = np.log(np.exp(1/(np.sin(x) + 1)) / (5/4 + 1/np.power(x, 15))) / np.log(1 + np.power(x, 2))
for i, k in zip(x, y):
    print("For x = {:4} y = {:.3f}".format(i, k))
```

# Упражнение №2

Постройте график функции

$ y(x) = x^2 - x - 6 $

и по графику найдите найдите корни уравнения $y(x) = 0$. (Не нужно применять численных методов — просто приблизьте график к корням функции настолько, чтобы было удобно их найти.)

```python
x = np.linspace(-5, 5.5, 100)
y = x*x - x - 6

# setting scale and labels
plt.ylim(-10, 15)
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$y$', fontsize=20)
plt.title(r'$y(x) = x^2 - x - 6$', fontsize=20)
#major grid lines
plt.grid(b=True, which='major', color='green', lw=1.5, ls='dashed')
# minor grid lines
plt.minorticks_on()
plt.grid(b=True, which='minor', ls='dashdot', color='gray', alpha=0.6)

plt.plot(x, y, c='red')
```

Из Графика можно увидеть что

$ y(x=-2) = 0 $

$ y(x=3) = 0 $

# Упражнение №3

Постройте график функции

$ y = \log_{1+\tan(\frac{1}{1+\sin(x)^2})}{(x^2+1)e^{-\frac{|x|}{10}}} $

```python
x = np.linspace(-100, 100, 300)
y = np.log((x*x + 1) * np.exp(-np.absolute(x) / 10)) / np.log(1 + np.tan(1 / (1 + np.sin(x)**2)))

plt.figure(figsize=(15, 5))
# setting labels
plt.xlabel(r'$x$', fontsize=20)
plt.title(r'$y = \log_{1+\tan(\frac{1}{1+\sin(x)^2})}{(x^2+1)e^{-\frac{|x|}{10}}} $', fontsize=25)
# major ticks grid
plt.grid(True, which='major', lw=1.5, ls='--')

plt.plot(x, y)
```

# Упражнение №5

Приблизить данные из приведённого примера с погрешностями или свои собственные (из лабораторного практикума по общей физике) многочленами первой и второй степени. Начертить точки с погрешностями и полученные аппроксимационные кривые на одном графике.

```python
plt.figure(figsize=(15, 8))
x = np.arange(1, 10, 1)
y = [0.99, 0.49, 0.35, 0.253, 0.18,
     0.09, -0.01, -.35, -0.8]

# estimating y errors
y_err = (y - np.round(y)) / 2

# plotting data points with errors
plt.errorbar(x, y, xerr=0.15, yerr=y_err, 
             label='Experiment data', ecolor='green', elinewidth=1.1)

# first degree polynomial estimation
p_1 = np.polyfit(x, y, deg=1, cov=True)[0]
p_f1 = np.poly1d(p_1)
plt.plot(x, p_f1(x), c='red', label="Fitted line(deg=1)", ls='--')

# second degree polynomial estimation
p_2 = np.polyfit(x, y, deg=2, cov=True)[0]
p_f2 = np.poly1d(p_2)
plt.plot(x, p_f2(x), c='orange', label='Fitted curve(deg=2)', ls='-.')

# label settings
plt.gca().tick_params(axis='both', labelsize=15)
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$y$', fontsize=20, rotation='horizontal', ha='right')
# gird settings
plt.grid(True)
plt.minorticks_on()
plt.grid(True, which='minor', ls='--', alpha=0.6)

plt.legend(loc='upper right', fontsize=15)
```

# Упражнение №6

Постройте график функции Вейерштрасса

${\displaystyle w(x)=\sum _{n=0}^{\infty }b^{n}\cos(a^{n}\pi x),}$


```python
plt.figure(figsize=(10, 6))
# Function parameters
a, b = 3, 0.5
X = np.linspace(0, 3, 1000)
Y = np.array([])
for x in X:
    # calculating y for each x
    Y = np.append(Y, sum([b**n * np.cos(a**n * np.pi * x) for n in range(10*6)]))

plt.gca().tick_params(axis='both', labelsize=15)
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$w(x)$', fontsize=20, rotation='horizontal', ha='right')
plt.title(r'$Weierstrass \indent function$', fontsize=20)
plt.grid(True)

plt.plot(X, Y)
```
